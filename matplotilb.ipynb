# **1. Matplotilb 이란?**

-   다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 파이썬 라이브러리
-   matplotlib을 이용하면 우리가 이전에 알아본 numpy나 pandas에서 사용되는 자료구조를 쉽게 시각화 할수있음

```
import numpy as np
import matplotlib.pyplot as plt
# 노트북 셀 내에 그림 출력
%matplotlib inline
# %matplotlib inline 의 역할은 notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것
# %pylab inline 또는 %matplotlib inline 을 실행하면 
#ipython 과 쥬피터 노트북에서 plt.show() 를 실행하지 않아도 되고 도표가 실행창에 포함된다
```

# **range 와 arange 차이점**

range는 step 정수값만 가능

arange는 numpy가 제공하는 기능으로써, 정수 실수 사용가능. 지정된 범위만큼 배열을 만들어주는것.



```
a= range(1,10)
b = np.arange(1,10)
print("a= {0}".format(a))
print("a = {0}".format(type(a)), end = "\n\n")

print("b = {0}".format(b))
print("b= {0}".format(type(b)))
```

# **plot()로 그래프를 그리기**

```
# sin그래프를 그림
s = np.sin(np.pi*np.arange(0.0, 4.0, 0.01)) # 총200개의 값이 나옴 
t = plt.plot(s)    # sin의 그래프를 그림
```

```
plt.plot(np.arange(0.0, 4.0, 0.01), s)  # x-축을 지정한다.
#x축을 지정할때 그릴려는 numpy변수의 범위와 같아야한다.
```

```
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.1))
plt.plot(np.arange(0.0, 2.0, 0.1), s1, 'o', ms = 4)
# 'o' 는 그래프를 점으로 찍어준다
#  ms=4는 점의 크기를 설정한다.
```

# 산포도 그리기

산포도란 대푯값을 중심으로 자료들이 흩어져 있는 정도를 의미한다. 이는 하나의 수치로서 표현되며 수치가 작을수록 자료들이 대푯값에 밀집되어 있고, 클수록 자료들이 대푯값을 중심으로 멀리 흩어져 있다.


# **정규분포(가우시안 표준 정규 분포)**

-    정규분포(normal distribution)는 가우시안 분포(Gaussian distribution)라고도 불린다. 평균에 가까울수록 발생할 확률이 높고 평균에서 멀어질수록 발생할 확률이 적은 현상
-   일한 확률을 가진 변수들이 적당히 크면, 해당 변수들의 평균은 일정한 패턴을 가짐

```
# 산포도(scatter plot)를 그린다
# 인수 마커를 사용하여 산포도를 작성
x = np.random.randn(5000)    
y = np.random.randn(5000)   

# randn은 기댓값이 0이고 표준편차가 1인 가우시안 표준 정규 분포를 따르는 난수를 생성

t = plt.plot(x, y, 'o', alpha=0.1)   #, ms=8)

# alpha는 투명도를 지정한다. 0이면 완전 투명, 1이면 완전 불투명
```

# **막대그래프 그리기**

-   막대그래프는 plot 가아니라 bar함수를 이용하여 그린다.
-   bar 차트는 plot과 달리 x값에 항목을 문자열로 구성된 리스로 넣는 게 아니라, 0부터 시작하는 정수 리스트로 넣어주어야 하므로 리스트의 길이를 range

```
# 막대그래프(bar graph)를 그린다.
means = [10,20,30]
stddevs = [1,0.7, 0.8]
# yerr는 오차범위를 지정해주고, 그안에 대입하는 값으로 크기를 조정
# 0이 제일작고 점점늘어남
bar_labels = ['bar 1', 'bar 2', 'bar 3']
# plot bars
x_pos = list(range(len(bar_labels)))
y_pos = list(range(10,50,10))
years = ['2017', '2018', '2019']
time = list(range(10,110,10))
plt.bar(x_pos, means, yerr=stddevs ,color = "pink", width =0.6) 
# yerr는 막대그래프위에 오차범위를 지정하는 함수
# bar함수는 맨처음으로 몇개의 값이 들어가는지 입력하고 그 다음 그래프를 그릴 값을 넣는다.
# color 는 그래프의 색을지정
# width 는 그래프의 두께를 지정
plt.xticks(x_pos,years) # x축값의 범위를지정
plt.yticks(y_pos,time) # y축값의 범위를 지정
plt.xlabel("Years",fontsize = 20)# x축라벨을 지정
plt.ylabel("Naumber",fontsize = 20) # y축라벨을 지정
plt.show()
```

# 히스토그램 그리기 

-   히스토그램은 연속형 자료에 대한 도수분포표를 시각화한 그래프
-   연속형 자료를 계급으로 나누어 계급별 도수를 막대로 나타냄

# *막대그래프와 히스토그램의 차이*

-  막대도표의 경우, 막대와 막대 사이에 간격이 벌어짐

   히스토그램에는 모든 막대가 공백 없이 서로붙어있음.


-  막대도표의 경우, 범주형 자료의 각 범주를 나타내므로 x축에 막대가 서로 떨어짐 

   스토그램의 경우, 연속형 자료의 값들이 모두 연속적으로 이어져 있기 때문

```
# 히스토그램 (histogram)
rng = np.random.RandomState(123)

# np.random.RandomState(123)는 랜덤넘버 생성기인 랜덤함수들을 포함하는 클래스
# (123)은 시드임. 시드란 어떤 특정한 시작 숫자를 정해주면 컴퓨터가 난수처럼 보이는 수열을생성.
# 이 시작값을 시드라고함.

x = rng.normal(0, 20, 1000) 
# normal 형식 = np.random.normal(loc=0.0, scale=1.0, size=None)
# loc = 평균 , scale = 표준편차

# fixed bin size
bins = np.arange(0, 30, 5) # fixed bin size 


plt.hist(x, bins=bins)
#히스토그램은  plt.hist(val, bins )로 그린다.
#bins는 구간을 설정.
plt.show()
```
