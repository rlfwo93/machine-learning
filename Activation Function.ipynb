```
import numpy as np
import matplotlib.pyplot as plt
```

# **활성화 함수(Activation Function)**

-  Data를 비선형으로 바꾸기위해 사용
-  데이터가 산형이면 망이 깊어지지않기 때문에 비선형으로 바꿈 .


# **망이 깊어지면 얻는 장점.**

- 매개 변수가 줄어든다.
- 필요한 연산의 수가 줄어든다.


# 활성화 함수 종류 

1. Sigmoid Function
- x 값이 작아질수록 0에 수렴하고, 커질수록 1에 수렴한다.

image.png

```
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5,5,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.0)
plt.show()
```

2. Step Function
- Step Function은 말 그대로 계단모양 함수로, 특정값 이하는 0, 이상은 1로 출력하도록 만들어진 함수이다.
함수 식은
y = 1 (when x≥0), 0 (when x<0) 와 같이 쓸 수 있다.

image.png

```
def step_function(x):
    return np.array(x>0.0, dtype = np.int)

x = np.arange(-5, 5, 0.001)
y = step_function(x)
plt.plot(x,y)
plt.show()
```

3. ReLU Function(Rectified Linear Unit)
- ReLU 함수는 입력이 특정값을 넘으면 입력이 그대로 출력으로 나오고, 0을 넘지 않을시 0을 반환하는 함수이다.
함수 식은
y = x (when x≥0), 0(when x<0)

image.png

```
def ReLU(x):
    return np.maximum(0,x)

x = np.arange(-5, 5, 0.1)
y = ReLU(x)

plt.plot(x,ReLU(x))
plt.show()
```
